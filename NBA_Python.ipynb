```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as po
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA

import json
import warnings
import urllib.request

pd.set_option('display.max_rows', 999)
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('rawdata.csv')
```

```python
win = win.drop(columns=['TEAM','W/L'])
```

```python
win
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(win)
```

```python
scaled = scaler.transform(win)
```

```python
pca = PCA(n_components=4)
```

```python
pca.fit(scaled)
```

```python
prcomp = pca.transform(scaled)
```

```python
prcomp = pd.DataFrame(data=prcomp)
```

```python
prcomp.head()
```

```python
fig = px.scatter(prcomp,x=0, y=1)
fig.show()
```

```python
pca.singular_values_
```

```python
pca.components_.T
```

```python
pca_res = pd.DataFrame(data = pca.components_.T, index=win.columns)
```

```python
scaled_pca = pd.DataFrame(data = scaled_pca)
```

```python
scaled = scaled_pca
```

```python
def elbow(X):
    sse = []
    
    for i in range(1,11):
        km = KMeans(n_clusters=i)
        km.fit(X)
        sse.append(km.inertia_)
        
    fig = px.line(sse)
    fig.show()
    
elbow(scaled)
```

```python
km = KMeans(n_clusters=3).fit(scaled)
```

```python
km.cluster_centers_
```

```python
predict = pd.DataFrame(km.predict(scaled))
```

```python
predict.columns = ['predict']
```

```python
scaled = pd.DataFrame(data=scaled)
```

```python
r = pd.concat([scaled, predict], axis=1)
```

```python
result = r
```
